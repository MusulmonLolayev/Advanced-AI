## Core Equations And Rules

Use these definitions consistently in your implementation.

- Dataset:

$$
D=\{(x_i,y_i)\}_{i=1}^{n}, \qquad x_i\in\mathbb{R}^d, \qquad y_i\in\{1,\dots,C\}
$$

- Class proportion at node $m$:

$$
p_{mc}=\frac{1}{n_m}\sum_{(x_i,y_i)\in D_m}\mathbf{1}\{y_i=c\}
$$

- Gini impurity:

$$
I_G(D_m)=1-\sum_{c=1}^{C}p_{mc}^{2}
$$

- Impurity decrease:

$$
\Delta I(j,t)=I(D_m)-\frac{|D_L|}{|D_m|}I(D_L)-\frac{|D_R|}{|D_m|}I(D_R)
$$
