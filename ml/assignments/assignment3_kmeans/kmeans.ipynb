## Core Equations And Rules

Use these definitions consistently in your implementation.

- Within-cluster sum of squares:

$$
J(c, \mu) = \sum_{r=1}^{k} \sum_{i:c_i=r} \|x_i - \mu_r\|_2^2
$$

- Assignment step:

$$
c_i = \arg\min_{r \in \{1, \dots, k\}} \|x_i - \mu_r\|_2^2
$$

- Centroid update:

$$
\mu_r = \frac{1}{|\{i:c_i=r\}|} \sum_{i:c_i=r} x_i
$$

- Standardization with training-set statistics only:

$$
x'_j = \frac{x_j - \mu_j}{\sigma_j}
$$

Implementation expectations:
- use squared Euclidean distance in the assignment step
- each update step should replace a centroid by the mean of its assigned points
- the inertia should not increase within one run of alternating updates
- multiple random restarts should be supported, and the best run should be retained
- if a cluster becomes empty, keep the behavior explicit and documented in your implementation

## Step 3: Verify one assignment-update cycle on a toy example

Use a very small dataset to check that your implementation assigns points to the nearest centroids, recomputes the means correctly, and reports the corresponding inertia.


```python
X = np.array([
    [1.0, 1.0],
    [1.0, 2.0],
    [5.0, 4.0],
    [6.0, 5.0],
])
centroids0 = np.array([
    [1.0, 1.0],
    [6.0, 5.0],
])

model = KMeansClustering(n_clusters=2, max_iter=10, n_init=1, random_state=231)
labels = model.assign_labels(X, centroids0)
centroids1 = model.update_centroids(X, labels, centroids0)
inertia1 = model.compute_inertia(X, labels, centroids1)

print("initial centroids:\n", centroids0)
print("labels:", labels)
print("updated centroids:\n", centroids1)
print("inertia:", round(float(inertia1), 6))
```
