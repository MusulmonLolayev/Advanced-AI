## Core Equations And Rules

Use these definitions consistently in your implementation.

- Dataset:

$$
X = \{x_1,\dots,x_n\}
$$

- Epsilon-neighborhood:

$$
N_\varepsilon(x_i) = \{x_j : \|x_j - x_i\|_2 \le \varepsilon\}
$$

- Core, border, and noise points:

$$
\text{core if } |N_\varepsilon(x_i)| \ge \text{MinPts}, \qquad
\text{border if it lies in a core neighborhood but is not core,} \qquad
\text{noise if it is neither core nor border.}
$$

- DBSCAN uses the parameters `eps` and `min_pts` in code, which correspond to `\varepsilon` and `MinPts` in the lecture.
