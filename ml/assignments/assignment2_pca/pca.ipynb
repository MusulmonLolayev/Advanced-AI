## Core Equations And Rules

Use these definitions consistently in your implementation.

- Mean and centered data:

$$
\mu = \frac{1}{n}\sum_{i=1}^{n}x_i, \qquad X_c = X - \mathbf{1}\mu^\top
$$

- Covariance matrix:

$$
\Sigma = \frac{1}{n-1}X_c^\top X_c
$$

- First principal direction:

$$
w_1 = \arg\max_{\|w\|_2=1} w^\top \Sigma w
$$

- Projection and reconstruction:

$$
z = W_k^\top (x-\mu), \qquad \hat{x} = \mu + W_k z
$$

- Reconstruction-error anomaly score:

$$
a(x) = \|x-\hat{x}\|_2^2
$$

- SVD route to PCA:

$$
X_c = U S V^\top, \qquad \Sigma = V \frac{S^2}{n-1} V^\top
$$

Implementation expectations:
- `covariance` and `svd` fitting should agree up to sign on the leading directions
- the principal directions should be orthonormal
- the reconstruction error should decrease as more components are kept
- fit on training data only when building downstream pipelines

## Step 3: Verify projection and reconstruction

Load the prepared warm-up matrix and confirm that your model centers the data, projects to one dimension, and reconstructs back to the original space.


````python
bundle = load_pca_core_bundle(require_file(PCA_CORE_DATA_PATH))
X = bundle["X"]
pca = PrincipalComponentAnalysis()
pca.fit(X, method="covariance")
print("mean:", np.round(pca.mean_, 4))
print("components:", np.round(pca.components_[:2], 4))
print("projection:", np.round(pca.transform(X, n_components=1), 4))
print("reconstruction:", np.round(pca.inverse_transform(pca.transform(X, n_components=1), n_components=1), 4))
````
